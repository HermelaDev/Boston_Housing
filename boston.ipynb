## Cleaning Data

```python
#checking for null values
boston.isnull().sum()
```

```python
# checking for duplicates
boston.duplicated().sum()
```

```python
# Renaming columns
boston.columns = ["crime_Rate",
    "zoned_Land",
    "industrial_Acres",
    "river_Proximity",
    "nitric_Oxides",
    "rooms_per_dwelling",
    "age_of_home",
    "employment_Distance",
    "highway_Accessibility",
    "property_Tax_Rate",
    "pupil_teacher_ratio",
    "black_Residents_Rate",
    "lower_Status_Population",
    "median_Home_Value"]

boston.tail(7)
```

```python
# Removing the columns that we won't need for this time 
boston = boston[["crime_Rate",
    "river_Proximity",
    "nitric_Oxides",
    "rooms_per_dwelling",
    "age_of_home",
    "employment_Distance",
    "highway_Accessibility",
    "property_Tax_Rate",
    "black_Residents_Rate"]]

boston.head(7)
```

```python
boston.dtypes
```

### Features Understanding

```python
tax = boston['property_Tax_Rate'].value_counts().head(10).plot(kind='bar', title='Tax Rates Count')

tax.set_xlabel('Property Tax Rate')
tax.set_ylabel('Count')
```

```python
NO = boston['nitric_Oxides'].plot(kind='kde', title='Nitric Oxides Distribution')

NO.set_xlabel('nitric_Oxides')
```

### Features Relationships

```python
boston.plot(kind = 'scatter', x='rooms_per_dwelling', y='property_Tax_Rate')
```

```python
boston.plot(kind = 'scatter', x='black_Residents_Rate', y='crime_Rate')
```

```python
sns.pairplot(boston, vars=["crime_Rate",
    "river_Proximity",
    "nitric_Oxides",
    "rooms_per_dwelling",
    "age_of_home",
    "employment_Distance",
    "highway_Accessibility",
    "property_Tax_Rate",
    "black_Residents_Rate"])
plt.show()
```

```python
boston_corr = boston[["crime_Rate",
    "river_Proximity",
    "nitric_Oxides",
    "rooms_per_dwelling",
    "age_of_home",
    "employment_Distance",
    "highway_Accessibility",
    "property_Tax_Rate",
    "black_Residents_Rate"]].corr()

boston_corr
```

```python
sns.heatmap(boston_corr)
```
